```python
raw.rename_channels(lambda s: s.strip('.'))
```

```python
events_from_annot, event_dict = mne.events_from_annotations(raw)
print(event_dict)
print(len(events_from_annot))
```

```python
custom_mapping = {'T0': 11, 'T1': 22, 'T3': 33}
(events_from_annot,
 event_dict) = mne.events_from_annotations(raw, event_id=custom_mapping)
print(event_dict)
print(events_from_annot[:5])
```

```python
rem_annot = mne.Annotations(onset=[5, 41],
                            duration=[16, 11],
                            description=['REM'] * 2)
raw.set_annotations(rem_annot)
(rem_events,
 rem_event_dict) = mne.events_from_annotations(raw, chunk_duration=1.5)
```
